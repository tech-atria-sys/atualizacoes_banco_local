```python
import shutil
import os
from datetime import datetime
import os
```

```python
class Rename():
    def __init__(self) -> None:
        self.now = datetime.today()
        # Usando f-strings com formatadores de data para simplificar o GenerateData
        self.data = self.now.strftime('%d-%m-%Y')
        self.tempo = self.now.strftime('%H%M')
        
        self.renomes = {
            'Posição': 'posição',
            'Posição - Renda Fixa': 'renda_fixa_coe',
            'Posição - Renda Variável': 'renda_variavel',
            'Saldo em CC (D 0)': 'saldo_cc',
            'NNM Detalhado': 'nnm',
            'Base BTG': 'base_btg',
            'Informações': 'informações',
            'Detalhado': 'rentabilidade',
        }

    def ValidarArquivos(self) -> bool:
        """Verifica se todos os arquivos de origem existem na pasta atual."""
        arquivos_faltantes = []
        for nome_original in self.renomes.keys():
            arquivo = f"{nome_original}.xlsx"
            if not os.path.exists(arquivo):
                arquivos_faltantes.append(arquivo)
        
        if arquivos_faltantes:
            print("Erro: Os seguintes arquivos não foram encontrados:")
            for arq in arquivos_faltantes:
                print(f"   - {arq}")
            return False
        
        print("Todos os arquivos encontrados. Iniciando processamento...")
        return True

    def RenameColumns(self):
        for original, novo in self.renomes.items():
            try:
                os.rename(f"{original}.xlsx", f"{novo}.xlsx")
            except FileNotFoundError:
                print(f"Aviso: {original}.xlsx não encontrado para renomear.")

    def CreateFolder(self):
        # O parâmetro exist_ok=True evita o erro se a pasta já existir
        caminho_base = f'C:\\Scripts\\relatórios\\{self.data}'
        caminho_final = f'{caminho_base}\\{self.tempo}'
        
        os.makedirs(caminho_final, exist_ok=True)
        return caminho_final

    def SendFiles(self, destino):
        for novo_nome in self.renomes.values():
            arquivo = f"{novo_nome}.xlsx"
            if os.path.exists(arquivo):
                os.replace(arquivo, os.path.join(destino, arquivo))
```

```python
# --- Execução ---
etl_arquivos = Rename()


# 1. Validação inicial
if etl_arquivos.ValidarArquivos():
    etl_arquivos.RenameColumns()
    caminho = etl_arquivos.CreateFolder()
    etl_arquivos.SendFiles(caminho)
    print("Processo finalizado.")
else:
    print("Cancelada devido a arquivos ausentes.")
```
